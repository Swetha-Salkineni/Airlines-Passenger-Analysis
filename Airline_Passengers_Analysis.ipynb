**Handling the Outliers**

This code handles outliers in the selected columns (Arrival Delay in Minutes, Departure Delay in Minutes, and Flight Distance) using the Interquartile Range (IQR) method.

```python
for i in ['Arrival Delay in Minutes','Departure Delay in Minutes','Flight Distance']:
  q1=dfmain[i].quantile(0.25)
  q3 = dfmain[i].quantile(0.75)
  iqr = q3-q1
  min = q1-1.5*iqr
  max = q3+1.5*iqr

  dfmain[i]=[min if x < min else max if x > max else x for x in dfmain[i]]
```

Identifies outliers using Q1 (25th percentile) and Q3 (75th percentile).

Caps extreme values:

If a value is below the lower bound, it is replaced with min (Q1 - 1.5 * IQR).

If a value is above the upper bound, it is replaced with max (Q3 + 1.5 * IQR).

Reduces the impact of extreme values, making the dataset more suitable for machine learning models.

Step 8: Outlier Treatment

This code visualizes the distribution of the selected numerical columns after applying outlier treatment. The boxplots help confirm whether extreme values have been effectively handled.

```python
for i in ['Arrival Delay in Minutes','Departure Delay in Minutes','Flight Distance']:
  sns.boxplot(dfmain[i])
  plt.title(f"{i}")
  plt.show()
```

If outliers were successfully treated, the boxplots should show fewer extreme points.

**Step 11: Splitting the Dataset**

Why Are We Using It?

The train-test split is essential for evaluating machine learning models. It helps ensure that the model is trained on one portion of the dataset and tested on another, preventing overfitting and providing a realistic measure of model performance. By setting aside 25% of the data for testing, we can check how well the model generalizes to unseen data.

```python
from sklearn.model_selection import train_test_split
x=dfmain.drop('satisfaction',axis = 1)
y= dfmain['satisfaction']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Results from the Above Code

The dataset is divided into:

75% Training Data (x_train, y_train) – Used to train the model.

25% Testing Data (x_test, y_test) – Used to evaluate model performance.

This split allows us to assess how well the model performs on unseen data before deploying it in a real-world scenario.

Setting random_state=42 ensures that every time we run the code, we get the same split, making results reproducible.

**Building a Logistic Regression Model**

Why Are We Using It?

Logistic Regression is a widely used classification algorithm that helps predict categorical outcomes, such as whether a passenger is satisfied or not. It estimates the probability that a given input belongs to a particular class and is especially useful for binary classification tasks.

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
```

Creates an instance of the LogisticRegression model.

The model learns patterns from the training data (x_train, y_train).



**Making Predictions with Logistic Regression**

Why Are We Using It?

After training the Logistic Regression model, we need to evaluate its performance. The predict() function is used to generate predictions on the test dataset (x_test). This helps us compare the model’s predicted values against actual values (y_test) to assess its accuracy.

```python
y_pred = regressor.predict(x_test)
```

Trains a logistic regression model to predict passenger satisfaction.

Generates Predictions: The trained model predicts the class labels (e.g., satisfied or not satisfied) for the test data.

Allows Performance Evaluation: These predictions can be compared with y_test using metrics like accuracy, precision, recall, F1-score, and AUC-ROC to measure how well the model performs.

**Accuracy Calculation**

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

The accuracy score measures how well the model correctly classifies test data, providing a simple performance metric where a higher accuracy indicates better predictions.

**Confusion Matrix Calculation**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

The confusion matrix helps evaluate the model’s performance by showing the number of correct and incorrect predictions for each class, helping to analyze misclassifications

**ROC Curve Analysis**

```python
from sklearn.metrics import roc_curve, auc
y_prob = regressor.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.show()
```

The ROC curve visualizes the trade-off between the true positive rate and false positive rate at different classification thresholds, while the AUC score quantifies the model’s ability to distinguish between classes, with higher values indicating better performance.

**AUC Score Calculation**

```python
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_test, y_prob)
print(f"AUC: {roc_auc:.2f}")
```

The AUC (Area Under the Curve) score measures the model's ability to distinguish between classes, where a score closer to 1 indicates a better-performing model.
